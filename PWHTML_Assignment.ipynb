2. Explain the purpose of comments in HTML and provide an example of how to use
   comments in an HTML documents?

   - Purpose of Comments in HTML-
   1. Clarifying the code: Comments help developers understand the purpose of specific sections of the code, making it easier to maintain or modify.

   2. Documenting the code: They can serve as a way to explain the logic, structure, or any special instructions to others or even to your future self.

   3. Temporarily disabling code: You can comment out sections of code that you want to ignore temporarily without deleting them.

3. Write an HTML program that includes a heading, a paragraph of text, a horizontal line, and a line break.
Arrange these elements to create a simple web page layout?


```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Simple Web Page</title>
</head>
<body>
    <h1>Welcome to My Web Page</h1>
    <hr> <!-- Horizontal line -->
    <p>This is a simple paragraph of text that provides some information about the page.</p>
    <br> <!-- Line break -->
    <p>Here is another paragraph after a line break.</p>
</body>
</html>
```

4. write a short note on Tag and Elements with an example?

  1. Tag-
 A tag is a part of an HTML document that defines the structure and behavior of web content. Tags are enclosed in angle brackets (< >) and usually come in pairs: an opening tag and a closing tag.

 Example-

5. what is the DOCTYPE declaration in HTML?

 -  The <!DOCTYPE> declaration in HTML is used to specify the document type and the version of HTML being used. It ensures that the web browser renders the page in standards mode rather than quirks mode. It must be placed at the very beginning of an HTML document, before the


```
<html>
```
tag.


```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Example</title>
</head>
<body>
    <h1>Hello, World!</h1>
</body>
</html>
```
